## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['cut','price','color']].groupby('cut').agg('mean').reset_index()
diamonds.head(30)

grafico_barras = sns.barplot(data=diamonds, x="cut", y="price")
grafico_barras.set(title="Gráfico com a média de valores dos diamantes por corte", xlabel = 'Tipo de corte', ylabel="Valores")
#grafico_barras.figure(figsize=(12,8))
```

**Insight do gráfico 1**:

De acordo com o gráfico acima, percebe-se que a média de valor do diamante modifica conforme o tipo de corte solicitado. Usando como exemplo, o corte "Premium" que possue uma média de valor acima do corte ideal

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = data.query('month == "Dec"')
flights.head()

with sns.axes_style('darkgrid'):
  grafico = sns.lineplot(data=flights, x='year', y='passengers', color='red')
  grafico.set(title="Número de passageiros anuais somente no mês de dezembro", xlabel='Ano', ylabel='Passageiros')

```

**Insight do gráfico 2**:

O gráfico revela um aumento anual no número de passageiros em dezembro ao longo dos anos, com exceção de um período entre 1957 e 1958, onde a variação não foi significativa

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year')
  grafico.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros')
```

**Insight do gráfico 3**:

Nos anos de 1957 e 1958, não houve diferença no número de passageiros em abril, setembro e provavelmente em dezembro, indicando estabilidade de um ano para outro. Em contraste, em 1960, houve um grande aumento contínuo no número de passageiros ao longo do ano, sem nenhum período onde o número fosse igual ao ano anterior
