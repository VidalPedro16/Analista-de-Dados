## 1\. Contexto do Projeto

O propósito deste empreendimento é ilustrar o processo de coleta de informações do Telegram, seguido pela sua transferência para um repositório central de dados, conhecido como DataLake. Após essa etapa, os dados são processados de forma batch na nuvem. O resultado desse processamento é uma base de dados limpa e estruturada, que permite a realização de análises detalhadas. Essas análises são cruciais para descobrir insights importantes, que podem ser utilizados para melhorar a qualidade dos serviços oferecidos e também para identificar e desenvolver novas oportunidades de geração de receita.

## 3\. Conheça a Ingestão




A etapa de ingestão em sistemas analíticos tem como função principal a captação dos dados brutos, mantendo-os o mais próximo possível de sua forma original, sem alterar seu conteúdo ou estrutura. Por exemplo, quando dados são recebidos de uma API web que opera com o protocolo REST, eles são armazenados no formato JSON, preservando assim a integridade da informação conforme foi fornecida.



O AWS API Gateway atua como um intermediário que aceita as mensagens recebidas pelo bot do Telegram, quando são enviadas através de um webhook, e dispara a execução de uma função específica no AWS Lambda.



Documentação: ([Amazon API Gateway](https://docs.aws.amazon.com/pt_br/apigateway/))

## 4\. Conheça o ETL

Na fase de ETL, os dados que já foram coletados e armazenados em sua forma bruta nos sistemas analíticos são processados. Isso inclui a extração desses dados, a aplicação de transformações necessárias para adequá-los às necessidades de análise e, finalmente, o carregamento em um formato e localização que facilitam a análise e a geração de insights.

Após a coleta, os dados brutos são submetidos a um processo contínuo de preparação, no qual são purificados, eliminadas as duplicações, entre outras limpezas. Em seguida, são armazenados utilizando métodos como particionamento, organização por colunas e compressão para otimizar o armazenamento e a performance. Com essas etapas concluídas, os dados estão aptos para serem utilizados em análises por especialistas na área.

Eu vou utilizar uma função no AWS Lambda como o motor de processamento e um bucket do AWS S3 para armazenar os dados processados como uma camada enriquecida. Para garantir que o processo seja recorrente, eu configurarei uma regra no AWS Event Bridge para acionar a função Lambda todos os dias.

O que é ETL: ([ETL](https://www.oracle.com/br/integration/what-is-etl/))

## 6\. Storytelling

A implementação de um pipeline de dados no Telegram, integrando-o com serviços AWS, demonstrou ser uma solução eficiente para coleta e análise de informações. A utilização da API do Telegram facilitou a extração de mensagens, enquanto a AWS Lambda permitiu automatizar o armazenamento seguro no Amazon S3. Este processo inicial foi essencial para estabelecer uma base sólida, que pode ser expandida para lidar com volumes maiores de dados e tipos de mídia mais complexos, conforme a necessidade do projeto.

Ao aplicar técnicas de particionamento, orientação a colunas e compressão, otimizamos o armazenamento dos dados no Amazon Athena, garantindo que as consultas sejam rápidas e eficientes. A automação diária do processo ETL, através do AWS EventBridge, assegura que o pipeline de dados permaneça atualizado e operante sem intervenção manual, o que é crucial para a escalabilidade e a continuidade do fluxo de informações. Por fim, a segurança dos dados é um benefício inestimável dessa integração, com a AWS oferecendo robustos protocolos de proteção. Pipelines de dados são fundamentais para consolidar informações provenientes de diversas origens, facilitando a sua transformação e preparação para análises. Ao eliminar a fragmentação de dados, esses pipelines asseguram que as análises sejam mais consistentes e exatas, promovendo uma visão unificada e confiável das informações, essencial para a tomada de decisão estratégica.
